```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
base_tempo = pd.read_csv("weather.nominal.csv")
```

```python
base_tempo
```

```python
x_play = base_tempo.iloc[:,0:4].values
x_play
```

```python
y_play = base_tempo.iloc[:,4].values
y_play
```

```python
from sklearn.preprocessing import LabelEncoder
le_outlook = LabelEncoder()
le_temperature =LabelEncoder()
le_humidity =LabelEncoder()
le_windy =LabelEncoder()
```

```python
x_play[:,0] = le_outlook.fit_transform(x_play[:,0])
x_play[:,1] = le_temperature.fit_transform(x_play[:,1])
x_play[:,2] = le_humidity.fit_transform(x_play[:,2])
x_play[:,3] = le_windy.fit_transform(x_play[:,3])
```

```python
x_play
```

```python
import pickle
with open('tempo.pkl','wb') as f:
    pickle.dump([x_play,y_play],f)
```

```python
naive_play = GaussianNB()
naive_play.fit(x_play,y_play)
```

```python
#rainy(1), cool(0), normal(1), true(1)

previsao = naive_play.predict([[1,0,1,1]])
```
